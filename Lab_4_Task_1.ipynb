```python
import math as mth
import random

# Base class: Geom
class Geom():
    geomType = 'Generic Geometry Type'

    def __init__(self, name=None, color=None):
        self.name = name if not name is None else random.choice(['Bill', 'Sally', 'Tamica', 'Josh', 'Lammar', 'Hussain'])
        self.color = color if not color is None else random.choice(['BLUE', 'RED', 'PURPLE', 'YELLOW'])

    def print_name(self):
        print('My name is', self.name, 'and my color is', self.color)

    @classmethod
    def area(self):
        return None

    def makeString(self):
        return f"Name: {self.name}, Color: {self.color}, Area: {self.area()}"

# Subclass: Circle
class Circle(Geom):
    def __init__(self, r):
        self.radius = r
        super().__init__()

    def area(self):
        return mth.pi * self.radius ** 2

# Subclass: Square
class Square(Geom):
    def __init__(self, s):
        self.side = s
        super().__init__()

    def area(self):
        return self.side ** 2

# New Subclass: Triangle
class Triangle(Geom):
    def __init__(self, base, height):
        self.base = base
        self.height = height
        super().__init__()

    def area(self):
        return 0.5 * self.base * self.height

# --- Testing the new Triangle class ---

# Create a triangle with base 6 and height 4
my_triangle = Triangle(6, 4)

# Print triangle properties and area
print('Base:', my_triangle.base, 'Height:', my_triangle.height)
my_triangle.print_name()
print('My area is', my_triangle.area())

# Create a list of triangles with base and height increasing
triangle_list = [Triangle(base, base+1) for base in range(3, 5)]

# Print all triangle info
for t in triangle_list:
    t.print_name()

# Use makeString to show formatted triangle info
print([t.makeString() for t in triangle_list])

# Optional - also test the existing Square and Circle
side = 8
my_square = Square(side)
print('Length of the side is', my_square.side)
my_square.print_name()
print('My area is', my_square.area())

circle_list = [Circle(i) for i in range(2, 4)]
print(circle_list)

for x in circle_list:
    x.print_name()

print([x.makeString() for x in circle_list])
```
